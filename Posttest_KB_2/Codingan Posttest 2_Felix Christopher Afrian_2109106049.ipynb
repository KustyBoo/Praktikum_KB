# Mengimport Library

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("CO2 Emissions_Canada.csv")
```

# Informasi yang di-dapat dan dimasukkan dalam dictionary

```python
info = {
  'total_records': 0,
  'attributes': [],
  'object_attributes': 0,
  'number_attributes': 0,
  'null_attributes': 0
}

info['total_records'] = df.shape[0]
info['attributes'] = list(df.columns.values)
info['null_attributes'] = list(df.loc[:, df.isnull().any()].columns)
info['object_attributes'] = list(df.select_dtypes(['object']).columns)
info['number_attributes'] = list(df.select_dtypes(['float64', 'int64']).columns)
```

# Jumlah Record

```python
print ("Jumlah Record :", info['total_records'] )
```

# Nama dan Jumlah Atribut

```python
print ("Nama-nama Attribute :", info['attributes'])
print ("Jumlah Attribute :", len(info['attributes']))
```

# Jumlah Atribut dengan Tipe Objek

```python
print ("Jumlah Atribut dengan Tipe Objek :", info['object_attributes'], len(info['object_attributes']))
```

# Jumlah Atribut dengan Tipe Angka

```python
print ("Jumlah Attribute dengan Tipe Angka :", info['number_attributes'], len(info['number_attributes']))
```

# Jumlah Atribut dengan Tipe Null

```python
print ("Jumlah Attribute yang Memiliki Nilai Null :", info['null_attributes'], len(info['null_attributes']))
```
