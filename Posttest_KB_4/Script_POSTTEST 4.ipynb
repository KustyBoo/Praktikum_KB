# Data Cleaning

### Mengecek apakah dataset memiliki nilai NULL

```python
dataset_split.isna().sum()
```

### Karena dataset awal tidak memiliki nilai NULL, kita import Dataset yang sudah kita rusak

```python
dataset_rusak = pd.read_csv("CO2 Emissions_Canada_Rusak.csv")
```

```python
dataset_rusak.head()
```

### Melihat jumlah dataset yang rusak per-atributnya

```python
dataset_rusak.isna().sum()
```

### Melihat semua kolom dan baris yang rusak dengan tabel

```python
dataset_rusak[dataset_rusak.isnull().any(axis=1)]
```

### Mengganti nilai NULL kolom bilangan bulat dengan median

```python
imputer = SimpleImputer(strategy='median')
```

### Memperbaiki kolom CO2 Emissions

```python
dataset_rusak['CO2 Emissions(g/km)'] = imputer.fit_transform(dataset_rusak[['CO2 Emissions(g/km)']])
```

```python
dataset_rusak.isna().sum()
```

```python
dataset_rusak.head()
```

### Mengganti nilai NULL kolom bilangan desimal dengan mean

### Memperbaiki kolom Fuel Consumption Comb, Fuel Consumption Hwy, dan Fuel Consumption City

```python
imputer = SimpleImputer(strategy='mean')
```

```python
dataset_rusak['Fuel Consumption Comb (L/100 km)'] = imputer.fit_transform(dataset_rusak[['Fuel Consumption Comb (L/100 km)']])
dataset_rusak['Fuel Consumption Hwy (L/100 km)'] = imputer.fit_transform(dataset_rusak[['Fuel Consumption Hwy (L/100 km)']])
dataset_rusak['Fuel Consumption City (L/100 km)'] = imputer.fit_transform(dataset_rusak[['Fuel Consumption City (L/100 km)']])
```

```python
dataset_rusak.isna().sum()
```

### Mengganti nilai NULL kolom tulisan dengan modus

### Memperbaiki kolom Make, Model, dan Vehicle Class

```python
imputer = SimpleImputer(strategy='most_frequent')
```

```python
dataset_rusak['Make'] = imputer.fit_transform(dataset_rusak[['Make']])
dataset_rusak['Model'] = imputer.fit_transform(dataset_rusak[['Model']])
dataset_rusak['Vehicle Class'] = imputer.fit_transform(dataset_rusak[['Vehicle Class']])
```

```python
dataset_rusak.isna().sum()
```

```python
dataset_rusak.head()
```

## Menangani Nilai Duplikat

```python
dataset_rusak[dataset_rusak.duplicated()]
```

```python
dataset_rusak.drop_duplicates(inplace=True)
```

```python
dataset_rusak[dataset_rusak.duplicated()]
```

```python
dataset_rusak.duplicated().sum()
```

# Mengganti tipe salah satu atribut angka

### Mengganti atribut CO2 Emissions dari float64 ke int64

```python
dataset_rusak['CO2 Emissions(g/km)'] = dataset_rusak['CO2 Emissions(g/km)'].astype('int64')
```

```python
dataset_rusak.dtypes
```

```python
dataset_rusak.head()
```

# Melakukan One-Hot Encoding

```python
onehot_encoder = OneHotEncoder(sparse=False)
```

```python
onehot1 = onehot_encoder.fit_transform(dataset_rusak[['Make']])
onehot2 = onehot_encoder.fit_transform(dataset_rusak[['Model']])
onehot3 = onehot_encoder.fit_transform(dataset_rusak[['Vehicle Class']])                                                     
```

```python
df_onehot1 = pd.DataFrame(onehot1)
df_onehot2 = pd.DataFrame(onehot2)
df_onehot3 = pd.DataFrame(onehot3)
```

```python
dataset_rusak = dataset_rusak.join(df_onehot1)
dataset_rusak = dataset_rusak.join(df_onehot2)
dataset_rusak = dataset_rusak.join(df_onehot3)
```
