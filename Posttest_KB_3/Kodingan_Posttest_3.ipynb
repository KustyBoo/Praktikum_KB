# Mengimport Library

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# Mengimport Dataset

```python
df = pd.read_csv("CO2 Emissions_Canada.csv")
```

```python
df.head()
```

# Visualisasi Data

## Grafik Batang

```python
X = list(df["Fuel Type"])
Y = list(df["CO2 Emissions(g/km)"])

plt.figure(figsize=(10,10))
plt.title("Emisi CO2 Gas/Bensin")
plt.xlabel("Jenis Bensin")
plt.ylabel("Tingkat Emisi")
plt.bar(X,Y, color="red")
plt.show()
```

### Keterangan :
#### Z = Premium gasoline
#### D = Diesel
#### X = Regular gasoline
#### E = Ethanol (E85)
#### N = Natural gas

### Kesimpulan  :
#### Gas Premium, Regular, dan Ethanol memiliki Emisi CO2 tertinggi dari semua jenis gas yang ada dengan tingkat rata rata emisi sekitar 400-500 g/km, sedangkan gas alami atau Natural Gas dan Diesel memiliki tingkat emisi terendah dengan rata rata emisi sekitar 200-300 g/km.

## ----------------------------------------------------------------------------------------------------------------------------------

## Grafik Pie

```python
data_pie = df["Fuel Type"].value_counts().rename_axis("Fuel Type").reset_index(name="ID_count")
```

```python
data_pie.head()
```

```python
plt.figure(figsize=(10,10))
plt.pie(
    data_pie["ID_count"], 
    labels=data_pie["Fuel Type"],
    autopct='%.2f%%',)
plt.legend()
plt.title("Penggunaan Bensin")
plt.show()
```

### Keterangan :
#### Z = Premium gasoline
#### D = Diesel
#### X = Regular gasoline
#### E = Ethanol (E85)
#### N = Natural gas

### Kesimpulan  :
#### Penggunaan gas Reguler dan gas Premium terhitung memakan presentase terbanyak dari semua tipe gas yang ada, 49,25% dari seluruh data menunjukkan bahwa gas reguler dipakai paling umum oleh masyarakat, sedangkan gas natural atau Natural Gas dipakai paling sedikit oleh masyarakat.

## Grafik Boxplot

```python
sns.boxplot(data=df,x='Fuel Type',y='Fuel Consumption City (L/100 km)')
plt.title("Penggunaan gas di perkotaan")
plt.show()
```

### Keterangan :
#### Z = Premium gasoline
#### D = Diesel
#### X = Regular gasoline
#### E = Ethanol (E85)
#### N = Natural gas

### Kesimpulan  :
#### Penggunaan gas dengan konsumsi terbanyak jatuh kepada Ethanol dengan penggunaan di sekitar median 20 L/100 km, dan tertinggi dengan jumlah sekitar 30 L/100 km, Serta penggunaan paling sedikit jatuh kepada Diesel dengan median sekitar 10 L/100 km dan tertinggi sekitar 14 L/100 km.
